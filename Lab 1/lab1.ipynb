### 1.Піднести до додатного цілого степеня дійсне ненульове число.

```python
def power(x, n):
    if n == 0:  
        return 1
    return x * power(x, n - 1) 

print(power(5, 2)) 
```

### 2.Рекурсивно знайти суму перших n натуральних чисел

```python
def sum_natural(n):
    if n == 1:  
        return 1
    return n + sum_natural(n - 1) 

print(sum_natural(5))  # 1 + 2 + 3 + 4 + 5 = 15
```

### 14.Виписати послідовність ходів для перекладання n дисків вежі (n = 2; 3; 4; 5 дисків, використати онлайн гру).

```python
def hanoi(n, source, auxiliary, target, pegs, step=[0]):
    if n == 1:
        step[0] += 1
        disk = pegs[source].pop()
        pegs[target].append(disk)
        print(f"Крок {step[0]}: Перемістити диск {disk} з {source} на {target}")
        print_pegs(pegs)
        return
    
    hanoi(n - 1, source, target, auxiliary, pegs, step)
    hanoi(1, source, auxiliary, target, pegs, step)
    hanoi(n - 1, auxiliary, source, target, pegs, step)

def print_pegs(pegs):
    print(f"A: {pegs['A']}, B: {pegs['B']}, C: {pegs['C']}")
    print("-" * 30)


n = 4
total_steps = 2**n - 1
print(f"Загальна кількість кроків для {n} дисків: {total_steps}\n")

pegs = {'A': list(range(n, 0, -1)), 'B': [], 'C': []}

print("Початковий стан:")
print_pegs(pegs)

hanoi(n, 'A', 'B', 'C', pegs)
```

```python
# алгоритм Ханойської вежі, але замість виведення дисків на кожному стержні виводить лише їх кількість
def hanoi(n, source, target, auxiliary, rods):
    if n == 1:
        move_disk(source, target, rods)
        return
    hanoi(n - 1, source, auxiliary, target, rods)
    move_disk(source, target, rods)
    hanoi(n - 1, auxiliary, target, source, rods)


def move_disk(source, target, rods):
    disk = rods[source].pop()
    rods[target].append(disk)
    print(f"Перемістити диск {disk} зі стрижня {source} на стрижень {target}")
    print_state(rods)


def print_state(rods):
    print(f"A = {len(rods['A'])}, B = {len(rods['B'])}, C = {len(rods['C'])}")
    print("-" * 30)




n = 4
rods = {
    'A': list(range(n, 0, -1)),  
    'B': [], 
    'C': []   
}

print(f"Початковий стан:")
print_state(rods)
hanoi(n, 'A', 'C', 'B', rods)

```
